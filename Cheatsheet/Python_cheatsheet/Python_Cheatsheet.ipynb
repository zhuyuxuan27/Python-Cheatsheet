## Plotting Cplor Setting

```python
def sinplot():
    x = np.linspace(0,14,100)
    plt.figure(figsize=(8,6))
    for i in range(4):
        plt.plot(x,np.sin(x+i)*(i+0.75),label = 'sin(x+i)*(i+0.75)')
    plt.legend()
```

```python
sinplot()
```

```python
with sns.axes_style('white'):
    sinplot()
```
